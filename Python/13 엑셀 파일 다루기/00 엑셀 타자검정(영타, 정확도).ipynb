```python
import pandas as pd

# 엑셀 파일 데이터프레임으로 저장하기
df = pd.read_excel('C:/myPyCode/data/play_typing2.xlsx', index_col = "학생")

# 원하는 값만 추출하기 위해서 데이터프레임 행과 열을 전치시키기(Transpose)
new_df = df.T

# 타수만 추출하기 위해서 짝수 데이터만 추출
typing = new_df[::2]
won_typing = typing["이도원"]
won_typing_list = won_typing.values.tolist()
won_data = {"영타": won_typing_list}
index_list = ["5일","6일","7일","12일","13일","14일","15일","16일","19일",
             "20일","21일","22일","23일","26일","27일","28일"]
column_list = ["영타"]
df_won_typing = pd.DataFrame(won_data, index=index_list, columns=column_list)

# 정확도만 추출하기 위해서 홀수 데이터만 추출
accuracy = new_df[1::2]
won_accuracy = accuracy["이도원"]
won_accuracy_list = won_accuracy.values.tolist()
won_data2 = {"정확도": won_accuracy_list}
index_list = ["5일","6일","7일","12일","13일","14일","15일","16일","19일",
             "20일","21일","22일","23일","26일","27일","28일"]
column_list = ["정확도"]
df_won_accuracy = pd.DataFrame(won_data2, index=index_list, columns=column_list)

# 타수 데이터와 정확도 데이터 합치기
df_won_typing.join(df_won_accuracy)

# 시각화를 위해서 인덱스를 따로 date 변수에 지정하기
date = df_won_typing.index


# 데이터 시각화하기
import matplotlib
import matplotlib.pyplot as plt

# 폰트 깨지지 않도록 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False

# 데이터 시각화를 위한 설정
%matplotlib inline

plt.plot(date, won_typing_list)
plt.xlabel("날짜")
plt.xticks(rotation=45)
plt.ylabel("영타")
plt.ylim(220, 300)
plt.title("10월 영타 - 이도원")
plt.grid(True)
plt.show()
plt.figure()

plt.plot(date, won_accuracy_list)
plt.xlabel("날짜")
plt.xticks(rotation=45)
plt.ylabel("정확도")
plt.ylim(96, 102)
plt.title("10월 영타 정확도 - 이도원")
plt.grid(True)
plt.show()
```
