```python
import pandas as pd
df = pd.read_excel('C:/myPyCode/data/play_typing2.xlsx', index_col = "학생")
new_df = df.T
new_df
```

```python
typing = new_df[::2]
typing
```

```python
won_typing = typing["이도원"]
won_typing
```

```python
won_typing_list = won_typing.values.tolist()
won_typing_list
```

```python
won_data = {"영타": won_typing_list}
index_list = ["5일","6일","7일","12일","13일","14일","15일","16일","19일",
             "20일","21일","22일","23일","26일","27일","28일"]
column_list = ["영타"]

df_won_typing = pd.DataFrame(won_data, index=index_list, columns=column_list)
df_won_typing
```

```python
accuracy = new_df[1::2]
accuracy
```

```python
won_accuracy = accuracy["이도원"]
won_accuracy
```

```python
won_accuracy_list = won_accuracy.values.tolist()
won_accuracy_list
```

```python
won_data2 = {"정확도": won_accuracy_list}
index_list = ["5일","6일","7일","12일","13일","14일","15일","16일","19일",
             "20일","21일","22일","23일","26일","27일","28일"]
column_list = ["정확도"]

df_won_accuracy = pd.DataFrame(won_data2, index=index_list, columns=column_list)
df_won_accuracy
```

```python
df_won_typing.join(df_won_accuracy)
```
