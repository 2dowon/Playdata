### 선생님 코드

```python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://www.naver.com")

html = req.text

soup = BeautifulSoup(html, "lxml")

my_titles = soup.select(
     "ul.list_nav"
    # "ul.type_fix li"
    # "ul.NM_FAVORITE_LIST li"
)
# print(my_titles)

title_result = []

for title in my_titles:
    title_result.append(title.text)
    
print(title_result)

import re

for text in title_result:
    result = re.sub("\n","",text)
    print(result)
```

#### 뒤쪽만

```python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://www.naver.com")

html = req.text

soup = BeautifulSoup(html, "lxml")

my_titles = soup.select(
    "ul.NM_FAVORITE_LIST li"
)
# print(my_titles)

title_result = []

for title in my_titles:
    title_result.append(title.text)
    
# print(title_result)

import re

for text in title_result:
    result = re.sub("\n","",text)
    print(result)
```
