```python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://www.hankookilbo.com/News/Read/A2020101618490003127")

html = req.text

soup = BeautifulSoup(html, "lxml")

print(soup.find('h2', {"class":"title"}).get_text(), "\n")

reads = soup.select("p.editor-p")

for read in reads:
    print(read.text)
```
